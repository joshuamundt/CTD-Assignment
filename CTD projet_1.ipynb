```python
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
```

SS=summer shallow;
WS=winter shallow

```python
SS=pd.read_csv("summershallow_B3.csv")
WS=pd.read_csv("wintershallow_B3.csv")
```

SD=summer deep ; WD= winter deep

```python
SD=pd.read_csv("summerdeep_B1.csv")
WD=pd.read_csv("winterdeep_B1.csv")
```

summer shallow

```python
A=SS.pressure
A1=SS.time
B=SS.seawater_temperature
C=SS.practical_salinity
```

z=depth [m]; z=h=P/(rho*g)

```python
z=A/(1000*9.81)
```

winter shallow

```python
D=WS.pressure
D2=WS.time
E=WS.seawater_temperature
F=WS.practical_salinity
```

```python
z1=D/(1000*9.81)
```

```python
z2=G
```

winter deep

```python
J=WD.pressure
J2=WD.time
K=WD.temp
L=WD.practical_salinity
```

```python
z3=J
```

```python
ssp=1449.2+(4.6*B)-(0.055*B**2)+(0.00029*B**3)+(1.34-0.01*B)*(C-35)+(0.016*z)
sspmean=np.mean(ssp)
sspmean
```

```python
plt.plot(z,ssp)
plt.ylabel('ssp [m/s]')
plt.xlabel('depth [m]')
plt.title('depth vs ssp')
plt.axhline(y=sspmean, color='r', linestyle='-')
```
